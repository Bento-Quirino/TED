# Árvores

Árvores são estruturas de dados que possuem uma raiz e nós. 

Diferente do nó de uma lista encadeada, que possui apenas um ponteiro para o próximo nó, o nó de uma árvore possui um ponteiro para cada um dos filhos.

As árvores são muito utilizadas em algoritmos de busca, como o algoritmo de busca em largura e o algoritmo de busca em profundidade.

```python
# Criando um dict para uma arvore binaria
import random

arvore = {
    'value': 'A',
    'leftNode': None,
    'rightNode': None
}

def createNode(value):
    return {
        'value': value,
        'leftNode': None,
        'rightNode': None
    }
    
def insertNode(node, nodeToInsert, side):
    if side == 'left':
        node['leftNode'] = nodeToInsert
    elif side == 'right':
        node['rightNode'] = nodeToInsert
    else:
        print('Invalid side')
        return False
    
    return True

def printTree(node):
    print(node['value'])
    if node['leftNode'] != None:
        print('leftNode:' + str(node['leftNode']['value']))
    else:
        print('leftNode: None')
    if node['rightNode'] != None:
        print('rightNode:' + str(node['rightNode']['value']))
    else:
        print('rightNode: None')
    print()
    
    if node['leftNode'] != None:
        printTree(node['leftNode'])
    if node['rightNode'] != None:
        printTree(node['rightNode'])
    
```

```python
# Criando nós para a arvore

arvore['leftNode'] = createNode('B')
arvore['rightNode'] = createNode('C')

# INdo para o nó B
arvore['leftNode']['leftNode'] = createNode('D')
arvore['leftNode']['rightNode'] = createNode('E')

# Indo para o nó C
arvore['rightNode']['leftNode'] = createNode('F')

# Inserindo um nó no nó F
arvore['rightNode']['leftNode']['leftNode'] = createNode('G')
arvore['rightNode']['leftNode']['rightNode'] = createNode('H')

print('Imprimindo a arvore')
printTree(arvore)

# Arvore com o formato

#     A
#    / \
#   B   C
#  / \  /
# D  E F
#     / \
#    G   H
```
